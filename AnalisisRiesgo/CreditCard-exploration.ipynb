```python
import scipy.stats as stats
# Referencia: Seaborn is a Python data visualization library based on matplotlib.
import seaborn as sns 

credit.columns

output = 'default payment next month'

# Exploratory Data Analysis
cols = [ f for f in credit.columns if credit.dtypes[ f ] != "object"]
# Removemos el ID y el default
cols.remove( "ID")
cols.remove( output )

f = pd.melt( credit, id_vars=output, value_vars=cols)
g = sns.FacetGrid( f, hue=output, col="variable", col_wrap=5, sharex=False, sharey=False )
g = g.map( sns.distplot, "value", kde=True).add_legend()
```

```python
# Observaciones:
# Default tienen una mayor proporciÃ³n de valores LIMIT_BAL inferiores
# Los no default tienen una mayor proporciÃ³n de mujeres (sexo = 2)
# Los no default tienen una mayor proporciÃ³n de mÃ¡s educados (EDUCATION = 1 o 2)
# Los no default tienen una mayor proporciÃ³n de solteros (MARRIAGE = 2)
# Los no default tienen una mayor proporciÃ³n de personas de 30 a 40 aÃ±os
# Los no default tienen una proporciÃ³n MUCHO mayor de variables PAY_X cero o negativas 
#    (esto significa que estar al dÃ­a o adelantarse a los pagos estÃ¡ asociado con no incumplir el mes siguiente). 
#    Esta es una relaciÃ³n fuerte ya que la distribuciÃ³n estÃ¡ mÃ¡s separada, por lo que esperamos que PAY_X sea importante.

# Definimos funcion ChiSquaredTestOfIndependence
#   DescripciÃ³n: The Chi-Square test of independence es usado para determinar si hay una relaciÃ³n significativa
#   entre dos variables nominales (categÃ³ricas). La frecuencia de cada categorÃ­a para una variable nominal es comparada
#   a travÃ©s de las categorias de la segunda variable nominal.

def ChiSquaredTestOfIndependence( df, inputVar, Outcome_Category ):
    # Es de utilidad tener esto en una funciÃ³n. El ChiSquaredTestofIndependence tiene una hipÃ³tesis nula: 
    # Outcome_Category es independiente de inputVar. Entonces creamos una estadÃ­stica de prueba que es una medida de 
    # la diferencia entre "esperado", es decir, lo que observarÃ­amos si Outcome_Category fuera independiente de inputVar
    # y "observado", es decir, lo que realmente muestran los datos. 
    # El valor p devuelto es la probabilidad de ver esta estadÃ­stica de prueba si la hipÃ³tesis nula es verdadera
    Outcome_Category_Table = df.groupby( Outcome_Category )[ Outcome_Category ].count().values
    Outcome_Category_Ratios = Outcome_Category_Table / sum( Outcome_Category_Table )
    possibleVals = df[inputVar].unique()
    observed = []
    expected = []
    for possible in possibleVals:
        countsInCategories = df[ df[ inputVar ] == possible ].groupby( Outcome_Category )[Outcome_Category].count().values
        if( len(countsInCategories) != len( Outcome_Category_Ratios ) ):
            print("Error! La clase " + str( possible) +" de \'" + inputVar + "\' no contiene todos los valores de \'" + Outcome_Category + "\'" )
            return
        elif( min(countsInCategories) < 5 ):
            print("Chi Squared Test necesita por lo menos 5 observaciones en cada celda!")
            print( inputVar + "=" + str(possible) + " tiene insuficientes datos")
            print( countsInCategories )
            return
        else:
            observed.append( countsInCategories )   
            expected.append( Outcome_Category_Ratios * len( df[df[ inputVar ] == possible ]))
    observed = np.array( observed )
    expected = np.array( expected )
    chi_squared_stat = ((observed - expected)**2 / expected).sum().sum()
    degOfF = (observed.shape[0] - 1 ) *(observed.shape[1] - 1 ) 
    
    p_value = 1 - stats.chi2.cdf(x=chi_squared_stat, df=degOfF)
    print("La prueba-estadistica calculada es %.2f" % chi_squared_stat )
    print("Si " + Outcome_Category + " es independiente de " + inputVar + ", este tiene una probabilidad %.2e de ocurrencia" % p_value )
    
```

```python
ChiSquaredTestOfIndependence( credit, "SEX", output )
```

```python
# Ok. "default" no es independiente de "SEX".
ChiSquaredTestOfIndependence( credit, "EDUCATION", output )  
```

```python
# No hay suficientes valores en Education class = 0. Tal vez encontremos lo mismo para los valores de Education = 4, 5, 6.

print("Tenemos %d con EDUCATION=0" % len(credit.loc[ credit["EDUCATION"]==0]))
print("Tenemos %d con EDUCATION=4" % len(credit.loc[ credit["EDUCATION"]==4]))
print("Tenemos %d con EDUCATION=5" % len(credit.loc[ credit["EDUCATION"]==5]))
print("Tenemos %d con EDUCATION=6" % len(credit.loc[ credit["EDUCATION"]==6]))
```

```python
# Ya que tenemos mas de 30mil registros, vamos a poner estas instancias no-tipicas de Education dentro de la clase EDUCATION=4 

credit["EDUCATION_Corr"] = credit["EDUCATION"].apply( lambda x: x if ((x>0) and (x<4)) else 4 )
ChiSquaredTestOfIndependence( credit, "EDUCATION_Corr", output ) 
cols.remove("EDUCATION")
cols.append("EDUCATION_Corr")

ChiSquaredTestOfIndependence( credit, "MARRIAGE", output ) 
```

```python
# Podemos ver que default no es independiente de EDUCATION_Corr ni independiente de MARRIAGE
# Separamos las variables en cuantitativas y cualitativas


# Variables cuantitativas:
quant = ["LIMIT_BAL", "AGE"]

# Variables cualitativas pero "codificadas" (la mayorÃ­a de ellas)
qual_Enc = cols
qual_Enc.remove("LIMIT_BAL")
qual_Enc.remove("AGE")
```

```python
# Con las variables PAY_, podemos ver que son importantes pero transformaremos las variables BILL_AMT y PAY_AMT 
# de NT Dollars a Log (NT Dollars)

logged = []
for ii in range(1,7): #rango de las variables es de 1 a 6
    qual_Enc.remove("PAY_AMT" + str( ii ))
    credit[ "log_PAY_AMT" + str( ii )]  = credit["PAY_AMT"  + str( ii )].apply( lambda x: np.log1p(x) if (x>0) else 0 )
    logged.append("log_PAY_AMT" + str( ii ) )

for ii in range(1,7):
    qual_Enc.remove("BILL_AMT" + str( ii ))
    credit[ "log_BILL_AMT" + str( ii )] = credit["BILL_AMT" + str( ii )].apply( lambda x: np.log1p(x) if (x>0) else 0 )
    logged.append("log_BILL_AMT" + str( ii ) )

f = pd.melt( credit, id_vars=output, value_vars=logged)
g = sns.FacetGrid( f, hue=output, col="variable", col_wrap=3, sharex=False, sharey=False )
g = g.map( sns.distplot, "value", kde=True).add_legend()
```

```python
# Parece que un registro mÃ¡s alto PAY_AMT estÃ¡ ligeramente asociado con un menor default.

# Entonces ahora tenemos variables cuantitativas, variables qual_Enc y variables logarÃ­tmicas. 
# Verifiquemos las correlaciones con la variable de salida:

features = quant + qual_Enc + logged + [output]
corr = credit[features].corr()
plt.subplots(figsize=(30,10))
sns.heatmap( corr, square=True, annot=True, fmt=".1f" ) 
```
